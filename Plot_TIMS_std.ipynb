```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data
srm915a = pd.read_csv('Data/SRM915a_std.csv')
srm915b = pd.read_csv('Data/SRM915b_std.csv')
```

```python
fig, ax = plt.subplots(1,1, figsize=(5, 3), sharey=True)

ax.scatter(srm915a.index, srm915a.d44Casw, ec='k', fc='green',
            s=100, label='SRM915a')
ax.axhline(y=np.mean(srm915a.d44Casw), ls='-', c='k')
ax.axhline(y=np.mean(srm915a.d44Casw)+np.std(srm915a.d44Casw), ls='--', c='k')
ax.axhline(y=np.mean(srm915a.d44Casw)-np.std(srm915a.d44Casw), ls='--', c='k')

ax.set_ylim(-2.08, -0.94)

ax.set_ylabel('$\delta^{44}Ca_{sw}$', fontsize=14)

ax1 = ax.twiny()
ax1.scatter(srm915b.index, srm915b.d44Casw, ec='k', fc='orange',s=100,
label='SRM915b', marker='s')
ax1.axhline(y=np.mean(srm915b.d44Casw), ls='-', c='k')
ax1.axhline(y=np.mean(srm915b.d44Casw)+np.std(srm915b.d44Casw), ls='--', c='k')
ax1.axhline(y=np.mean(srm915b.d44Casw)-np.std(srm915b.d44Casw), ls='--', c='k')
ax1.set_ylim(-2.08, -0.94)

fig.legend(loc='center')


ax1.tick_params(axis='x', labeltop=False, top=False, bottom=False)
ax.tick_params(axis='x', labelbottom=False, bottom=False)

plt.tight_layout()
plt.savefig('Figure/SRM915_Plot.png', dpi=300)
```
